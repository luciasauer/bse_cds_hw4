```python
import pandas as pd
from diabetes_detector.preprocessing.pp_functions import fill_na_with_mean, make_binary_gender, drop_missing_values_in_subset
from diabetes_detector.train_test_model.tt_functions import  apply_one_hot_encode, split_dataset, train_rfc, calculate_roc_auc_score, keep_only_important_features
```

```python
df = pd.read_csv("sample_diabetes_mellitus_data.csv")
df.sample(10)
```

```python
df = drop_missing_values_in_subset(df, ["age", "gender", "ethnicity"])
X_train, X_test, y_train, y_test = split_dataset(df)
```

```python
for col in ['height', 'weight']:
    X_train = fill_na_with_mean(X_train, col)
    X_test = fill_na_with_mean(X_test, col)
```

```python
X_train = apply_one_hot_encode(X_train, "ethnicity")
X_test = apply_one_hot_encode(X_test, "ethnicity")
```

```python
X_train = make_binary_gender(X_train)
X_test = make_binary_gender(X_test)
```

```python
X_train = keep_only_important_features(X_train)
X_test = keep_only_important_features(X_test)
```

```python
model = train_rfc(X_train, y_train)
```

```python
y_pred_train = model.predict_proba(X_train)
y_pred_test = model.predict_proba(X_test)
```

```python
calculate_roc_auc_score(y_train, y_pred_train)
```

```python
calculate_roc_auc_score(y_test, y_pred_test)
```
